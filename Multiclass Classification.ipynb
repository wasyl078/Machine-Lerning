### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Data generatar

```python
# all data
def standard_data_generator(N):
    
    x_data = np.random.uniform(0, 0.5, size = N)
    y_data = np.random.uniform(0, 0.5, size = N)
    
    return {'x1': random_data(1.5, 0.3, N), 'y1': random_data(1.5, 0.55, N), 
            'x2': random_data(0.5, 0.45, N), 'y2':random_data(1.5, 0.45, N),
            'x3': random_data(0.5, 0.55, N), 'y3': random_data(0.5, 0.35, N),
            'x4': random_data(1.5, 0.45, N), 'y4': random_data(0.5, 0.2, N)}
    
def random_data(a, rng, N):
    data = np.random.uniform(a - rng, a + rng, size = N)
    return np.array([[d] for d in data])
```

### How does the data actually look like

```python
# settings
x_min = 0
x_max = 2.01
y_min = 0
y_max = 2.01
period = 0.1

# graph settings
plt.figure(figsize=(9,9), dpi=100)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(x_min,x_max)
plt.ylim(y_min, y_max)
plt.xticks(np.arange(x_min, x_max, period))
plt.yticks(np.arange(y_min, y_max, period))

# raw data call:
data = standard_data_generator(15)
plt.plot(data['x1'], data['y1'], 'x', color='red')
plt.plot(data['x2'], data['y2'], 'x', color='green')
plt.plot(data['x3'], data['y3'], 'x', color = 'blue')
plt.plot(data['x4'], data['y4'], 'x', color = 'yellow')

# showing resluts
plt.title("Raw Data", fontsize = 15)
plt.show()
```
